```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
rain = pd.read_csv("D:/DataSets for ML/Rainfall.csv")
```

```python
rain.head()
```

```python
rain.isnull().mean()*100
```

```python
rain['winddirection'] = rain['winddirection'].fillna(rain['winddirection'].mean())
```

```python
rain['windspeed'] = rain['windspeed'].fillna(rain['windspeed'].mean())
```

```python
rain.isnull().mean()*100
```

```python
rain.sample(5)
```

```python
rain.info()
```

```python
X = rain.drop(columns=['rainfall','day'])
Y = rain['rainfall']
```

```python
X
```

```python
Y
```

```python
Y = Y.map({'yes': 1, 'no': 0})
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_train_scaled
```

```python
x_test_scaled = scaler.transform(x_test)
x_test_scaled
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=150,
    learning_rate=0.01,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb.fit(x_train, y_train)

y_predxgb = xgb.predict(x_test)

accuracy_score(y_test, y_predxgb)
```

```python
import joblib

joblib.dump(xgb, 'xgboost_rain_model.pkl')
```
